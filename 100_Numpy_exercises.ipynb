#### 32. Is the following expressions true? (★☆☆)
```python
np.sqrt(-1) == np.emath.sqrt(-1)
```

````python
# np.sqrt(-1) == np.emath.sqrt(-1) → False

# En matemáticas de números reales:
# La raíz cuadrada de -1 NO existe porque no hay número real cuyo cuadrado sea negativo
# Por eso: np.sqrt(-1) = nan

# En matemáticas de números complejos:
# Se inventó la unidad imaginaria: i² = -1
# Entonces: √(-1) = i
# Por eso: np.emath.sqrt(-1) = 1j (1j es i en Python)

# emath = Extended Math (matemáticas extendidas a complejos)

# Ejemplos:
# np.sqrt(-1) = nan
# np.emath.sqrt(-1) = 1j
# np.emath.sqrt(4) = (2+0j)
# np.emath.sqrt(-4) = 2j  → porque (2j)² = 4j² = 4(-1) = -4

# Respuesta: False (nan != 1j)
print(np.sqrt(-1) == np.emath.sqrt(-1))  # False
````

#### 38. Consider a generator function that generates 10 integers and use it to build an array (★☆☆)

````python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

arr = np.fromiter(fibonacci(10), dtype=int)

print(arr)
````

#### 48. Print the minimum and maximum representable value for each numpy scalar type (★★☆)

````python
for dtype in [np.int8, np.int16, np.int32, np.int64, np.uint8, np.uint16, np.uint32, np.uint64, np.float32, np.float64]:
    if np.issubdtype(dtype, np.integer):
        info = np.iinfo(dtype)
    else:
        info = np.finfo(dtype)
    print(f"{dtype.__name__}: min={info.min}, max={info.max}")
````
